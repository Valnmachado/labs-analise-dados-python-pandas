```python
# Lista a quantidade de meses com dados existente por ano no DataFrame, para saber se entre eles possuem 
# a mesma quantidade de meses, para a comparação de dados anuais.
#

df_fato.loc[:,['Mês','Ano']].sort_values('Mês').groupby(by=['Ano','Mês']).apply(list)
```

```python
# função para cálculo do preço médio em dolar, tanto para ANO ou MÊS E ANO.
#

def usmedio(vlfob, kgliquido):
    return (vlfob / kgliquido)*1000

def mes_ext(mes):
    mes_ext = ['Unknown','JAN','FEV','MAR','ABR','MAI','JUN','JUL','AGO','SET','OUT','NOV','DEZ']
    return mes_ext[mes]

#----------------------


# Consulta os volumes das exportação de morango por ANO
#

df_FatoAno = df_fato.loc[:,[ 'Ano',
                             'Quilograma Líquido',
                             'Valor FOB (US$)'
                           ]
                         ].groupby(['Ano']).sum()


display(df_FatoAno.groupby(['Ano']).sum().style.format( { 'Quilograma Líquido' :'{:,.0f}',
                                                          'Valor FOB (US$)'    :'{:,.0f}'
                                                         }
                                                       )
       )

#----------------

# Consulta os volumes das exportação de morango por ANO E MÊS
#

df_FatoAno_Mes = df_fato.loc[:,[ 'Mês',
                                 'Ano',
                                 'Quilograma Líquido'
                               ]
                             ].groupby(['Ano','Mês']).sum()


display(df_FatoAno_Mes.groupby(['Ano','Mês']).sum().style.format( { 'Quilograma Líquido' :'{:,.0f}',
                                                                    'Valor FOB (US$)'    :'{:,.0f}'
                                                                   }
                                                                )
       )

#----------------
```

```python
# Volumes Exportado de Morango pelo o Brasil por Ano e Blocos, em ordenado por Quilograma Líquido do maior para o menor.
#

filtro = (df_fato.Ano >= 2018) & (df_fato.Ano <=2021)

#---------------

df_fatoPais = df_fato.loc[filtro,[ 'Ano',
                                   'Países',
                                   'Bloco Econômico',
                                   'Quilograma Líquido',
                                   'Valor FOB (US$)'
                                 ]
                          ].groupby(['Ano','Bloco Econômico']).sum()
#---------------

#df_fatoPais['US$/TONELADAS'] = usmedio(df_fatoPais['VL_FOB (US$)'], df_fatoPais['KG_LIQUIDO'])

df_fatoPais.sort_values(['Ano','Quilograma Líquido'], ascending=False).style.format({'Quilograma Líquido':'{:,.0f}',
                                                                                     'Valor FOB (US$)':'{:,.0f}'})
#---------------                                    
```

```python
# Volumes Exportado de Morango pelo o Brasil por Ano e Países.
#

df_fato.groupby([df_fato['Ano'],'Países'])['Quilograma Líquido'].sum().reset_index() \
                                                                .style.format({'Quilograma Líquido':'{:,.0f}',
                                                                               'Valor FOB (US$)'   :'{:,.0f}'})
```

```python
#lista os 10 Países com maiores volumes importados de morango de 2018 a 2021.
#

df_fato.groupby([df_fato['Ano'],'Países'])['Quilograma Líquido'].sum() \
                                                                .reset_index() \
                                                                .nlargest(10,['Quilograma Líquido']) \
                                                                .style.format({'Quilograma Líquido':'{:,.0f}',
                                                                               'Valor FOB (US$)':'{:,.0f}'})
```

```python
# Alterando o estilo do gráfico 
#

plt.style.use("ggplot")
```

```python
# Gráfico por ano e volumes exportado de morango pelo Brasil.
#

df_Exp_Ano = df_fato.groupby(df_fato['Ano'])['Quilograma Líquido'].sum().reset_index()

df_Exp_Ano = df_Exp_Ano.groupby(df_Exp_Ano['Ano'])['Quilograma Líquido'].sum()

#------------

df_Exp_Ano.plot.bar(title = "Exportação em Quilograma de Morango por Ano")
plt.xlabel("Ano")
plt.ylabel("Quilograma Líquido")
plt.xticks(rotation='horizontal');

#-------------------
# Exibe os volumes plotados no gráfico no formato de tabela.
#-------------------

df_Exp_Ano.reset_index().style.format({'Quilograma Líquido':'{:,.0f}'})
```

```python
# Gráfico por ano e valor em US$ Fob exportado de morango pelo Brasil.
#

df_Exp_Ano = df_fato.groupby(df_fato['Ano'])['Valor FOB (US$)'].sum().reset_index()

df_Exp_Ano = df_Exp_Ano.groupby(df_Exp_Ano['Ano'])['Valor FOB (US$)'].sum()

#------------

df_Exp_Ano.plot.bar(title = "Exportação em Valor FOB (US$) de Morango por Ano", color='red')
plt.xlabel("Ano")
plt.ylabel("Valor FOB (US$)")
plt.xticks(rotation='horizontal');

#-------------------
# Exibe os valores em US$ por ano plotados no gráfico no formato de tabela.
#-------------------

df_Exp_Ano.reset_index().style.format({'Valor FOB (US$)':'{:,.0f}'})
```
